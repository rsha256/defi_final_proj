```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

pd.set_option('display.max_columns', 50)

# Default plot configurations
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['figure.dpi'] = 150
sns.set()

import warnings
warnings.filterwarnings("ignore")

from IPython.display import display, Latex, Markdown

# get data from csv
pancakeswap = pd.read_csv('pancakeswap.csv', index_col=False)
pancakeswap
```

```python
label = "TxnFee(USD)"
pancakeswap.sort_values(label, ascending=True)
```

```python
label = "TxnFee(USD)"
tbl = pancakeswap
ax = sns.lineplot(x=tbl['DateTime'], y=tbl[label], label=label)
ax.set(xticklabels=[])
plt.legend(loc='upper left')
x = "DateTime"
plt.xlabel(x)
plt.ylabel(label)
plt.title(f"{label} vs {x}");
```

```python
label = "TxnFee(USD)"
series = pancakeswap[label]
series
```

```python
ax = sns.kdeplot(data=pancakeswap, y="TxnFee(USD)", shade=True, vertical=True)
ax.set(xlim=(0, 3))
plt.title("PancakeSwap: Kernel Density Estimate Plot");
```
