```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Default plot configurations
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['figure.dpi'] = 150
sns.set()

import warnings
warnings.filterwarnings("ignore")

from IPython.display import display, Latex, Markdown

# get data from csv
curvedata, uniswap, sushiswap = "curvedata", "uniswap", "sushiswap"
curvedata = pd.read_csv(f'etherscan-{curvedata}.csv')
uniswap = pd.read_csv(f'etherscan-{uniswap}data.csv')
sushiswap = pd.read_csv(f'etherscan-{sushiswap}data.csv')
curvedata
```

```python
tbl = curvedata.rename({"TxnFee(ETH)":"curvedata TxnFee(ETH)"})
sns.lineplot(x=tbl['DateTime'], y=tbl["TxnFee(ETH)"], label="curvedata TxnFee(ETH)")

tbl = uniswap.rename({"TxnFee(ETH)":"uniswap TxnFee(ETH)"})
sns.lineplot(x=tbl['DateTime'], y=tbl["TxnFee(ETH)"], label="uniswap TxnFee(ETH)ta")

tbl = sushiswap.rename({"TxnFee(ETH)":"sushiswap TxnFee(ETH)"})
sns.lineplot(x=tbl['DateTime'], y=tbl["TxnFee(ETH)"], label="sushiswap TxnFee(ETH)")

plt.legend(loc='upper left')
plt.xlabel("DateTime")
plt.ylabel("TxnFee (ETH)")
plt.title("TxnFee(ETH) vs DateTime");
```
