```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Default plot configurations
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['figure.dpi'] = 150
sns.set()

import warnings
warnings.filterwarnings("ignore")

from IPython.display import display, Latex, Markdown

# get data from csv
curvedata, uniswap, sushiswap = "curvedata", "uniswap", "sushiswap"
curvedata = pd.read_csv(f'etherscan-{curvedata}.csv')
uniswap = pd.read_csv(f'etherscan-{uniswap}data.csv')
sushiswap = pd.read_csv(f'etherscan-{sushiswap}data.csv')
curvedata
```

```python
tbl = curvedata
sns.lineplot(x=tbl['DateTime'], y=tbl["TxnFee(ETH)"], label="TxnFee(ETH)")
plt.legend(loc='upper left')
plt.xlabel("DateTime")
plt.ylabel("TxnFee (ETH)")
plt.title("TxnFee(ETH) vs DateTime");
```

```python
# outlier detection
label = "TxnFee(ETH)"
for _ in range(8):
    sushiswap = sushiswap.loc[sushiswap[label] != sushiswap[label].max()]
sushiswap.sort_values(label, ascending=False)
```

```python
label = "TxnFee(ETH)"
cd_label = f"curvedata {label}"
tbl = curvedata.rename({label:cd_label})
sns.lineplot(x=tbl['DateTime'], y=tbl[label], label=cd_label)

us_label = f"uniswap {label}"
tbl = uniswap.rename({label:us_label})
sns.lineplot(x=tbl['DateTime'], y=tbl[label], label=us_label)

ss_label = f"sushiswap {label}"
tbl = sushiswap.rename({label:ss_label})
sns.lineplot(x=tbl['DateTime'], y=tbl[label], label=ss_label)

plt.legend(loc='upper left')
x = "DateTime"
plt.xlabel(x)
plt.ylabel(label)
plt.title(f"{label} vs {x}");
```

```python
filename = 'data.txt'
data = open(filename).read().splitlines()
# print(data)
n = 1000
assert n == len(data), "error reading all data"
split_point = n >> 1
spent, received = np.array(data[::2]), np.array(data[1::2])
# print(spent, received, sep="\n"*3)
spent, received = spent.astype(np.float), received.astype(np.float)
total_ETH_spent = np.cumsum(spent)
ETH = 1e18
total_ETH_spent /= ETH
WBTC = 1e8
received /= WBTC
received **= 2

print(total_ETH_spent, received,sep='\n'*3)
```

```python
sns.lineplot(total_ETH_spent, received)
x, y = "Total ETH Spent", "Received"
plt.xlabel(x)
plt.ylabel(y)
plt.title(f"{y} vs {x}");
```
