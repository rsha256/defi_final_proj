```python
filename = 'data.txt'
data = open(filename).read().splitlines()
n = 1000
assert n == len(data), "error reading all data"
riffle = lambda deck: [deck[x // 2 + (x % 2)*(len(deck) // 2)] for x in range(len(deck))]
riffled = riffle(data)
split_point = n >> 1
spent, received = np.array(riffled[:split_point]).astype(np.float), np.array(riffled[split_point:]).astype(np.float)
print(spent, received)
```

```python
total_ETH_spent = np.cumsum(spent)
conversion_rate = 60_000
total_ETH_spent /= conversion_rate
received /= conversion_rate
print(total_ETH_spent, received)
```
